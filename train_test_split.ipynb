```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load dữ liệu
df = pd.read_csv("dataset/raw/US_Accidents_March23.csv")
print(f"Kích thước dataset gốc: {df.shape}")
print(f"\nThông tin cơ bản:")
print(df.info())
```

## Train-Test Split
**Random Split** - Chia ngẫu nhiên, phù hợp cho phân tích tổng quát

```python
def stratified_train_test_split(df, target_col='Severity', stratify_cols=['Severity'], test_size=0.2, random_state=42):
    train, test = train_test_split(df, test_size=test_size, random_state=random_state, stratify=df[stratify_cols])

    print(f"Train set: {train.shape[0]:,} records ({train.shape[0]/len(df)*100:.1f}%)")
    print(f"Test set:  {test.shape[0]:,} records ({test.shape[0]/len(df)*100:.1f}%)")

    print("\n📊 Phân phối {} (Train):".format(target_col))
    print(train[target_col].value_counts())
    print(f"Tỷ lệ: {train[target_col].value_counts(normalize=True) * 100}")

    print("\n📊 Phân phối {} (Test):".format(target_col))
    print(test[target_col].value_counts())
    print(f"Tỷ lệ: {test[target_col].value_counts(normalize=True) * 100}")

    return train, test
```

```python
train, test = stratified_train_test_split(df)
```

```python
train.to_csv("dataset/raw/train.csv", index=False)
test.to_csv("dataset/raw/test.csv", index=False)
```
